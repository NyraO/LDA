```python
def charNormalization(text):
    h2=['ріђ', 'ріЂ', 'ріѓ', 'ріЃ', 'ріё', 'ріЁ', 'ріє']
    h1=['рѕђ', 'рѕЂ','рѕѓ', 'рѕђ', 'рѕё', 'рѕЁ', 'рѕє']
    h3=['рѕљ', 'рѕЉ', 'рѕњ', 'рѕЊ', 'рѕћ', 'рѕЋ', 'рѕќ']
    t1=['рїИ','рї╣', 'рї║', 'рї╗', 'рї╝', 'рїй', 'рїЙ']
    t2=['рЇђ','рЇЂ', 'рЇѓ', 'рЇЃ', 'рЇё', 'рЇЁ', 'рЇє']
    s2=['рѕа','рѕА','рѕб','рѕБ','рѕц','рѕЦ','рѕд']
    s1=['рѕ░','рѕ▒', 'рѕ▓','рѕ│','рѕ┤','рѕх','рѕХ']
    a2=['рІљ', 'рІЉ', 'рІњ', 'рІЊ', 'рІћ', 'рІЋ', 'рІќ']
    a1=['ріа', 'ріА', 'ріб', 'ріа', 'ріц', 'ріЦ', 'рід']
    for i in range(len(h1)):
        text=text.replace(h3[i],h1[i])
        text=text.replace(h2[i],h1[i])
        text=text.replace(t2[i],t1[i])
        text=text.replace(s2[i],s1[i])
        text=text.replace(a2[i],a1[i])
    return text
```

```python
def stopWordRemoval(tokens):
    token_new=[]
    for x in tokens:
        if x not in normalised_stopWords:
            token_new.append(x)
    return token_new
```

def calc_term_frequence(stemmed_term_list):
    dict1={}
    freqList=[]
    term=[]
    for x in stemmed_term_list:
        freqList.append(stemmed_term_list.count(x))
        term.append(x)
    for x,y in zip(term,freqList):
        dict1[x]=y
    return dict1
